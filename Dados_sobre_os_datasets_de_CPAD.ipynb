## Criação do atributo idade no dataset Candidatos

```python
import pandas as pd
from datetime import datetime

# Carregar o dataset
df = pd.read_excel('dataset-candidatos.xlsx')  # Substitua pelo caminho do seu arquivo

# Converter a coluna de data de nascimento para o tipo datetime
df['DT_NASCIMENTO'] = pd.to_datetime(df['DT_NASCIMENTO'], errors='coerce')

# Função para calcular a idade
def calcular_idade(data_nascimento):
    if pd.isna(data_nascimento):  # Verifica se a data é válida
        return None
    hoje = datetime.now()
    return hoje.year - data_nascimento.year - ((hoje.month, hoje.day) < (data_nascimento.month, data_nascimento.day))

# Aplicar a função para calcular a idade
df['idade'] = df['DT_NASCIMENTO'].apply(calcular_idade)

# Salvar o DataFrame modificado em um novo arquivo Excel
df.to_excel('novo_arquivo.xlsx', index=False)  # Substitua pelo caminho desejado
```

```python
import pandas as pd

# Carregar os datasets
df_eleitorado = pd.read_excel('eleitorado.xlsx')
df_candidatos = pd.read_excel('candidatos.xlsx')

print(df_eleitorado.columns)

# Mapeamento de grau de instrução
mapa_instrucao = {
    'ANALFABETO': 0,
    'LE E ESCREVE': 1,
    'ENSINO FUNDAMENTAL INCOMPLETO': 2,
    'ENSINO FUNDAMENTAL COMPLETO': 3,
    'ENSINO MEDIO INCOMPLETO': 4,
    'ENSINO MEDIO COMPLETO': 5,
    'SUPERIOR INCOMPLETO': 6,
    'SUPERIOR COMPLETO': 7,
    'NAO INFORMADO': None  # Ou use np.nan
}

# Aplicar o mapeamento ao dataset do eleitorado
df_eleitorado['DS_GRAU_ESCOLARIDADE'] = df_eleitorado['DS_GRAU_ESCOLARIDADE'].map(mapa_instrucao)

# Remover linhas com valores NaN após o mapeamento (opcional, dependendo do tratamento desejado)
df_eleitorado = df_eleitorado.dropna(subset=['DS_GRAU_ESCOLARIDADE'])

# Calcular a média ponderada do grau de instrução do eleitorado por estado (SG_UF)
media_instrucao_eleitorado = (
    df_eleitorado.groupby('SG_UF').apply(
        lambda x: (x['DS_GRAU_ESCOLARIDADE'] * x['QT_ELEITORES_PERFIL']).sum() / x['QT_ELEITORES_PERFIL'].sum()
    )
).reset_index(name='MEDIA_ESCOLARIDADE')

# Exibe a quantidade de eleitores utilizados no cálculo (ou seja, o número de eleitores por estado)
quantidade_eleitores = df_eleitorado.groupby('SG_UF')['QT_ELEITORES_PERFIL'].sum()
print("Quantidade de eleitores utilizados por estado:")
print(quantidade_eleitores)

# Filtra apenas os candidatos que foram eleitos no dataset dos candidatos
candidatos_eleitos = df_candidatos[df_candidatos['DS_SIT_TOT_TURNO'] == 'ELEITO']

# Mapeamento de grau de instrução dos candidatos
candidatos_eleitos['DS_GRAU_INSTRUCAO'] = candidatos_eleitos['DS_GRAU_INSTRUCAO'].map(mapa_instrucao)

# Remover candidatos com grau de instrução não informado
candidatos_eleitos = candidatos_eleitos.dropna(subset=['DS_GRAU_INSTRUCAO'])

# Calcular a média do grau de instrução dos candidatos eleitos por estado (SG_UF) sem ponderação
media_instrucao_candidatos = (
    candidatos_eleitos.groupby('SG_UF')['DS_GRAU_INSTRUCAO']
    .mean()  # Calcula a média simples, sem ponderação
    .reset_index(name='MEDIA_ESCOLARIDADE_CANDIDATOS')
)

# Mescla os dois DataFrames (do eleitorado e dos candidatos) com base no estado (SG_UF)
correlacao_df = media_instrucao_eleitorado.merge(media_instrucao_candidatos, on='SG_UF')

# Exibir o DataFrame resultante
print(correlacao_df)
```

```python
# Cálculos de correlação entre as médias
correlacao_pearson = correlacao_df[['MEDIA_ESCOLARIDADE', 'MEDIA_ESCOLARIDADE_CANDIDATOS']].corr(method='pearson')
correlacao_spearman = correlacao_df[['MEDIA_ESCOLARIDADE', 'MEDIA_ESCOLARIDADE_CANDIDATOS']].corr(method='spearman')
correlacao_kendall = correlacao_df[['MEDIA_ESCOLARIDADE', 'MEDIA_ESCOLARIDADE_CANDIDATOS']].corr(method='kendall')

# Imprimir os resultados das correlações
print("Correlação de Pearson:")
print(correlacao_pearson)

print("\nCorrelação de Spearman:")
print(correlacao_spearman)

print("\nCorrelação de Kendall:")
print(correlacao_kendall)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Organizando as correlações em um DataFrame
correlacao_data = {
    'Pearson': [correlacao_pearson.iloc[0, 1]],
    'Spearman': [correlacao_spearman.iloc[0, 1]],
    'Kendall': [correlacao_kendall.iloc[0, 1]]
}

# Criando um DataFrame para as correlações
correlacao_df = pd.DataFrame(correlacao_data, index=['Correlação'])

# Plotando o heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlacao_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar=True)
plt.title('Correlação entre média de escolaridade do eleitorado e dos eleitos')
plt.show()
```
